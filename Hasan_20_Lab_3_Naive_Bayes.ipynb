```python
import pandas as pd
```

```python
data = pd.read_csv("weather.nominal.csv")
```

```python
data
```

```python
y = data['play']
y
```

```python
x = data.drop('play', axis = 'columns')
x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
x.outlook = le.fit_transform(data['outlook'])
```

```python
x.temperature = le.fit_transform(data['temperature'])
```

```python
x.humidity = le.fit_transform(data['humidity'])
```

```python
x.windy = le.fit_transform(data['windy'])
```

```python
x
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(x,y)
```

```python
predicted = model.predict([[1,2,0,1]])
```

```python
print("predicted value : ",predicted)
```
